##**Outliers Values**

Dari hasil persentase outlier, hampir semua kolom punya data yang menyimpang dalam jumlah kecil (sekitar 2–6%), tetapi Trans Fat (g) paling banyak memiliki nilai yang terlalu tinggi atau terlalu rendah dibanding data lainnya (10.89%), disusul Cholesterol (mg) (7.32%), sehingga dua kolom ini perlu diperhatikan karena bisa mempengaruhi hasil analisis.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

##**Eksplorasi Data (EDA)**

###**Comparison/Perbandingan**

Pada bagian **Perbandingan** ini, akan terlihat perbandingan nutrisi antar perusahaan fast food. Seperti yang sudah terlihat pada bar chart **Average Calories per Company**, kita bisa membandingkan rata-rata kalori yang ditawarkan oleh setiap perusahaan. Ini membantu kita mengidentifikasi perusahaan mana yang cenderung punya menu dengan kalori lebih tinggi (seperti Burger King) atau lebih rendah (seperti KFC), memberikan gambaran awal untuk pengembangan menu yang lebih sehat atau pilihan bagi konsumen.

```python
calories_per_company = df.groupby('Company')['Calories'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=calories_per_company.index, y=calories_per_company.values, palette='viridis', hue=calories_per_company.index, legend=False)
plt.title('Average Calories per Company')
plt.xlabel('Company')
plt.ylabel('Average Calories')
plt.xticks(rotation=45)
plt.show()
```

###**Composition/Komposisi**

Pada bagian Komposisi ini, hasil analisis bagaimana elemen-elemen dalam dataset terdistribusi atau membentuk keseluruhan. Melalui pie chart **Proporsi Jumlah Item Berdasarkan Perusahaan**, kita dapat melihat dengan jelas porsi item menu yang disediakan oleh setiap perusahaan fast food di dataset ini. Ini menunjukkan bahwa McDonald’s menyumbang porsi terbesar, diikuti oleh KFC, Burger King, Taco Bell, Wendy’s, dan Pizza Hut, memberikan gambaran visual tentang kontribusi masing-masing perusahaan terhadap total item yang tercatat.

###**Relationship/Hubungan**

Pada bagian **Relationship/Hubungan** ini, analisis bertujuan untuk mengidentifikasi bagaimana variabel-variabel dalam dataset saling terkait. Dengan heatmap **Korelasi antara Kalori dan Total Lemak** yang baru saja ditampilkan, terlihat adanya korelasi positif yang sangat kuat (nilai korelasi sekitar 0.96) antara jumlah kalori dan total lemak. Ini mengindikasikan bahwa makanan dengan kandungan lemak yang lebih tinggi cenderung memiliki jumlah kalori yang lebih tinggi pula, sebuah hubungan yang penting untuk memahami profil nutrisi pada menu fast food.

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Calories', 'Total Fat\n(g)']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Kalori dan Total Lemak')
plt.show()
```
