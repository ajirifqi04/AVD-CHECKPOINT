##Business Understanding

**Business Objective**

Dataset ini berisi informasi nutrisi dari berbagai menu makanan yang tersedia di restoran fast food, dengan kolom yang mencakup berbagai jenis informasi nutrisi seperti jumlah kalori, total lemak, kolesterol, natrium, karbohidrat, serat, gula, protein, serta poin Weight Watchers. Setiap baris mewakili satu item menu dari perusahaan tertentu, seperti McDonald’s, yang memberi gambaran tentang komposisi gizi setiap item makanan mereka.

Restoran fast food seringkali dikaitkan dengan makanan yang kurang sehat karena tingginya jumlah kalori, lemak, dan natrium dalam menu mereka. Banyak konsumen yang ingin menjaga pola makan sehat namun sulit untuk membuat keputusan yang tepat di restoran cepat saji, terutama karena sulit untuk membandingkan nilai gizi antara berbagai menu. Selain itu, perusahaan-perusahaan fast food perlu memastikan bahwa mereka menawarkan pilihan menu yang memenuhi standar kesehatan atau dapat disesuaikan dengan preferensi konsumen yang peduli dengan kesehatan.

Tujuannya adalah untuk memberikan wawasan tentang komposisi gizi menu fast food guna membantu perusahaan dalam merancang strategi menu yang lebih sehat, meningkatkan kepuasan pelanggan, dan menarik konsumen yang peduli kesehatan, sekaligus mendukung pengambilan keputusan yang lebih tepat terkait penawaran produk yang memenuhi kebutuhan diet spesifik pelanggan.

**Assess Situation**

Dataset ini memberikan informasi nutrisi dari berbagai menu makanan cepat saji, namun terdapat beberapa tantangan yang perlu diperhatikan dalam menganalisisnya. Pertama, dataset ini mungkin terbatas pada menu dari restoran tertentu (seperti McDonald’s), sehingga tidak mewakili seluruh variasi menu di industri restoran fast food. Kedua, tidak ada informasi mengenai ukuran porsi, yang bisa mempengaruhi total kalori dan nilai gizi. Selain itu, dataset ini hanya mencakup beberapa aspek gizi penting dan tidak mencakup mikronutrien atau alergi makanan, yang membatasi pemahaman tentang kesehatan makanan tersebut. Keterbatasan lainnya adalah bahwa data ini hanya fokus pada menu makanan, tanpa mempertimbangkan preferensi konsumen, pola makan, atau demografi, yang penting untuk menganalisis kepuasan pelanggan. Terakhir, data yang terbatas pada satu sumber restoran dan potensi variasi dalam resep antar lokasi dapat mengurangi representativitasnya untuk analisis yang lebih luas.


**Analytic Goals Project Plan**

Tujuan dari analisis ini adalah untuk menganalisis hubungan antara komposisi gizi dan kepuasan pelanggan, dengan fokus pada mengidentifikasi pola antara nilai gizi seperti kalori, lemak, dan protein terhadap rating pelanggan. Selain itu, analisis ini bertujuan untuk mengidentifikasi menu yang lebih sehat dan memberikan rekomendasi untuk pengembangan menu yang lebih sesuai dengan kebutuhan konsumen yang peduli kesehatan.

Langkah pertama adalah memahami dan memeriksa struktur dataset, kemudian membersihkan data dari nilai yang hilang atau tidak konsisten. Selanjutnya, eksplorasi data akan dilakukan untuk menganalisis hubungan antara komposisi gizi dan rating pelanggan. Setelah itu, segmentasi menu berdasarkan kategori gizi dilakukan untuk menemukan pola sehat, dan akhirnya, rekomendasi akan disusun untuk membantu restoran merancang menu yang lebih sehat dan meningkatkan kepuasan pelanggan.
